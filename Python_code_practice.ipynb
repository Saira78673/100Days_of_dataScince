# **NestedLoop**

```python
# 22. Write a nested loop to print this pattern
for i in range(1,5):
  for j in range(i):
    print('*' , end='')
  print()
```

```python
rows = int(input("enter number of rows "))
for i in range (1,rows):
  for j in range (1, i+1):
    print('*',end=' ')
  print()
```

```python
# Pattern 2
row=int(input("enter rows"))
for i in range(1,row+1):
  for j in range(1,i+1):
    print(j,end='')
  for k in range(i-1,0,-1):
    print(k,end='')
  print()
```

```python
# 23. Write a loop that goes through the list [12, 7, 18, 3, 25, 9] and prints only numbers greater than 10, but stops completely as soon as it finds a number greater than 20.
numbers = [12, 7, 18, 3, 25, 9]
for num in numbers:
  if num>20:
    break
if num > 10:
    print(num)
```

```python
#  Check if a number is prime using for...else:
n=7
for i in range(2,n):
  if i%2==0:
    print(n,'The number is not prime')
    break
  else:
    print(n,' prime')
```

# **BreakStatment**

```python
# A programm to check prime number
lower = int(input("enter lower range"))
upper = int(input('enter upper range'))
for i in range(lower, upper+1):
  for j in range(2,i):
    if i%j==0:
      break
  else:
    print(i)
```

```python
# write a python programm to covert a string to tiltle case without using title()
# 'hi how are you'.title( )
s=input('Enter sting : ')
l=[]
for i in s.split():

  l.append(i[0].upper() + i[1:].lower())
print('  '.join(l))
```
