```python

# Remove All Negative Numbers
n = [-1,4 ,-3,5,9]
positive =[]
for num in n:
  if num >= 0:
    positive.append(num)
print(positive)
```

```python
# Seperate Even and Odd numbers
n=[1,2,3,4, 5,6,7]
even=[]
odd=[]
for i in n:
  if i%2==0:
    even.append(i)
  else:
     odd.append(i)

print('Even:',even)
print('Odd',odd)
```

```python
# 8. Check if List is Palindrome
n = [ 1,2,3,2,1]
if n == n[::-1]:
  print('Palindrome')
else:
  print('not palindrome')
```
