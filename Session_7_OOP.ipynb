# **Mini program 1 : ATM withdrwal **

```python
class Atm():
    # constructor
    def __init__(self):
        self.pin = ' '
        self.balance = 0
        self.manu()

    # functionality
    def manu(self):
        user_input = input("""
      Hi! how can i help you?
      1.press 1 to generete pin
      2.press 2 to change pin
      3.press 3 to check balance
      4.press 4 to withdraw amount
      5.press 5 to exit
         """)
        print(user_input)
        if user_input =='1':
            # create pin
            self.creat_pin()
        elif user_input == '2':
            # change pin
            self.change_pin()
        elif user_input == '3':
            # check balance
            self.check_balance()
        elif user_input =='4':
            # withdraw amount
            self.withdraw()
        else:
            exit()

    # adding functionality
    def creat_pin(self):
        user_pin = input('Enter your pin: ')
        self.pin = user_pin
        user_balance = int(input('enter your balance: '))
        self.balance = user_balance
        print("Pin is created successfully...")
        self.manu()

    # change pin function
    def change_pin(self):
        old_pin = input('Enter your old pin: ')
        if old_pin == self.pin:
            # let him to change pin
            new_pin = input("Enter your new pin: ")
            self.pin = new_pin
            print("Pin change successfully!")
            self.manu()
        else:
            print('we are sorry!')
            self.manu()

    # Balance Check function
    def check_balance(self):
        user_pin = input('Enter your pin: ')
        if user_pin == self.pin:
            print('Your balance is:', self.balance)
            self.manu()
        else:
            print('Wrong password!')
            self.manu()

    # Withdraw amount
    def withdraw(self):
        user_pin = input('Enter your pin! ')
        if user_pin == self.pin:
            # allow withdrawal
            amount = int(input('Enter your amount: '))
            if amount <= self.balance:
                self.balance -= amount
                print("withdrawal successfully, balance is:", self.balance)
                self.manu()
            else:
                print("sorry! Insufficient balance")
                self.manu()
        else:
            print("Its not your account")
            self.manu()

# Triggering the object creation outside the class (unindented)
obj = Atm()

# Triggering the object creation outside the class (unindented)
obj = Atm()









```

# **Mini programm 2 : student class**

```python
# Create a class named Student
class Student:

    # Constructor
    # It runs automatically whenever an object is created.
    def __init__(self, name, age, marks):

        # Store the values inside the object.
        self.name = name
        self.age = age
        self.marks = marks

    # Method to display student information.
    def display(self):

        print("Student Name :", self.name)
        print("Age          :", self.age)
        print("Marks        :", self.marks)


# Create the first student object.
student1 = Student("Ali", 20, 87)

# Create the second student object.
student2 = Student("Sara", 21, 93)

# Display both students.
student1.display()

print("----------------")

student2.display()



```

#**Snake Water GUN game in python**

```python
import random

import random  # Yeh line zaroori hai taake random.randint kaam kare

def check(comp, user):
    if comp == user:
        return 0

    if (comp == 0 and user == 1):
        return -1

    if (comp == 1 and user == 2):  # Yahan space ka masala tha, ab theek hai
        return -1

    if (comp == 2 and user == 0):  # Isko bhi sahi se align kar diya hai
        return -1

    return 1  # Agar upar koi condition match nahi hui, toh user jeet gaya

comp = random.randint(0, 2)
user = int(input("0 for Snake, 1 for water and 2 for GUN:\n"))

score = check(comp, user)

print("You:", user)
print("Computer:", comp)

if (score == 0):
    print("its a draw")
elif (score == -1):
    print("You lose")
else:
    print("You won")
```

```python
from abc import ABC, abstractmethod # Abstraction ke liye

# ================== 1. FUNCTION ==================
# Function: code ka reusable block
def calculate_gpa(marks_list): # Parameter
    """Ye function list ke marks ka GPA nikalta hai"""
    total = sum(marks_list) # Built-in function use
    gpa = total / len(marks_list) / 10 # Simple formula
    return round(gpa, 2) # return value

# ================== 2. CLASS & OBJECT ==================
class Person: # Base Class
    university_name = "FAST University" # Class Variable - sab objects ke liye common

    def __init__(self, name, age): # Constructor - object bante hi chalta hai
        self.name = name # Instance Variable - har object ka alag
        self.age = age

    def display(self): # Instance Method
        print(f"Name: {self.name}, Age: {self.age}")

# ================== 3. INHERITANCE TYPES ==================

# 3.1 SINGLE INHERITANCE: 1 Parent -> 1 Child
class Student(Person): # Student inherits Person
    def __init__(self, name, age, roll_no):
        super().__init__(name, age) # Parent ka constructor call
        self.roll_no = roll_no
        self.courses = [] # Aggregation ke liye list

    def display(self): # Method Overriding - Polymorphism
        super().display() # Parent wala bhi call
        print(f"Roll No: {self.roll_no}")

# 3.2 MULTILEVEL INHERITANCE: A -> B -> C
class GraduateStudent(Student): # Student se inherit
    def __init__(self, name, age, roll_no, thesis_topic):
        super().__init__(name, age, roll_no)
        self.thesis_topic = thesis_topic

    def display(self): # Phir se Override
        super().display()
        print(f"Thesis: {self.thesis_topic}")

# 3.3 MULTIPLE INHERITANCE: 2 Parents -> 1 Child
class Sports:
    def play_sports(self):
        print("Playing Cricket")

class Researcher:
    def do_research(self):
        print("Doing Research")

class PhDStudent(GraduateStudent, Sports, Researcher): # 3 classes se inherit
    def __init__(self, name, age, roll_no, thesis_topic, stipend):
        GraduateStudent.__init__(self, name, age, roll_no, thesis_topic) # Multiple parent init
        self.stipend = stipend

# 3.4 HIERARCHICAL INHERITANCE: 1 Parent -> Multiple Children
class Teacher(Person): # Person se inherit
    def __init__(self, name, age, subject):
        super().__init__(name, age)
        self.subject = subject

    def display(self): # Override
        super().display()
        print(f"Subject: {self.subject}")

# 3.5 HYBRID INHERITANCE: Multiple + Multilevel ka mix
# PhDStudent upar hybrid hai: Multilevel + Multiple

# ================== 4. AGGREGATION ==================
# Aggregation: "Has-A" relationship. Class ke andar dusri class ka object
class Course: # Independent class
    def __init__(self, course_name, credits):
        self.course_name = course_name
        self.credits = credits

    def info(self):
        print(f"Course: {self.course_name}, Credits: {self.credits}")

# Student class mein already self.courses = [] hai
# Matlab Student "has-a" list of Course objects
def add_course_to_student(student, course): # Function to show aggregation
    student.courses.append(course) # Student ke andar Course ka object add
    print(f"{course.course_name} added to {student.name}")

# ================== 5. POLYMORPHISM ==================
# Polymorphism: "Ek naam, kai faces" - Same method name, different behavior

# 5.1 Method Overriding: upar display() har class mein alag hai
# 5.2 Duck Typing: function kisi bhi object ko accept karega jisme method ho
def print_details(person_obj): # Koi bhi Person, Student, Teacher de do
    person_obj.display() # Jis class ka object hoga uska display chalega
    print("---")

# ================== 6. ABSTRACTION ==================
# Abstraction: Sirf zaroori cheezein dikhao, detail chupao
# ABC = Abstract Base Class. Iska object nahi ban sakta
class Account(ABC): # Abstract Class

    @abstractmethod
    def calculate_fee(self): # Abstract Method - body nahi hai, child ko banana padega
        pass

    def show_account_type(self): # Normal Method
        print("This is a University Account")

class FeeAccount(Account): # Child must implement abstract method
    def __init__(self, base_fee, scholarship):
        self.base_fee = base_fee
        self.scholarship = scholarship

    def calculate_fee(self): # Abstract method ko implement kiya
        return self.base_fee - self.scholarship

# ================== 7. ENCAPSULATION ==================
class Bank: # Data hiding example
    def __init__(self, balance):
        self.__balance = balance # Private variable - __ se shuru

    def deposit(self, amount): # Public method se access
        self.__balance += amount

    def get_balance(self): # Getter
        return self.__balance

# ================== MAIN CODE - TESTING ==================
if __name__ == "__main__":
    print("1. FUNCTION TEST:")
    gpa = calculate_gpa([80, 90, 75]) # Function call
    print(f"GPA: {gpa}\n")

    print("2. SINGLE + MULTILEVEL INHERITANCE + POLYMORPHISM:")
    s1 = Student("Ali", 20, "S101")
    g1 = GraduateStudent("Sara", 24, "G202", "AI in Healthcare")
    print_details(s1) # Same function, different output
    print_details(g1)

    print("3. MULTIPLE INHERITANCE:")
    p1 = PhDStudent("Usman", 28, "P303", "Quantum Computing", 50000)
    p1.display()
    p1.play_sports() # Sports class ka method
    p1.do_research() # Researcher class ka method
    print()

    print("4. HIERARCHICAL INHERITANCE:")
    t1 = Teacher("Dr. Ahmed", 45, "Python")
    print_details(t1)

    print("5. AGGREGATION:")
    c1 = Course("OOP", 3) # Course object
    c2 = Course("DBMS", 3)
    add_course_to_student(s1, c1) # Student ke andar course add
    add_course_to_student(s1, c2)
    for course in s1.courses: # Student ke andar wale courses
        course.info()
    print()

    print("6. ABSTRACTION:")
    # acc = Account() # ERROR: Abstract class ka object nahi ban sakta
    fee = FeeAccount(100000, 20000)
    fee.show_account_type()
    print(f"Final Fee: {fee.calculate_fee()}\n")

    print("7. ENCAPSULATION:")
    b = Bank(5000)
    # print(b.__balance) # ERROR: Private hai
    b.deposit(2000)
    print(f"Balance: {b.get_balance()}") # Method se access
```
