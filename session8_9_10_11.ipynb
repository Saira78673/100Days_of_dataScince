# **File Handling using read()**

```python
# creating dempo file

f = open("demo.txt",'w')
f.write("Line1: Hellow\nLine2: World\nLine3: python")
f.close()

# Reading form the file
f = open("demo.txt",'r')
d =f.read()

print(d)
print("Type:" ,type(d))  # checking data type
f.close()


```

# **File Handling using Readline**

```python
f = open('demo.txt','r')
print("file reading using readline ")
Line1 = f.readline()
print(Line1,end='')
Line2 = f.readline()
print(Line2,end='')
Line3 = f.readline()
print(Line3,end='')
f.close
```

# **Using context manager with** automatically open and close file

```python
with open('demo.txt','r') as f :
  data = f.read()
  print("using with:" ,data)
  print("File close automatically")
```

# **tell() gives current pointer position. seek() moves pointer.**

```python
f= open("demo.txt",'r')
print("Current Position:" ,f.tell())
data = f.read(10)
print(data, f.tell())
f.seek(0)
print("Position after seek(0)" , f.tell())
f.close()
```

# **Working with binary**

```python
# write in binary mode

with open("data.bin","wb") as f:
  f.write(b"Hello Binary\n")
  f.write(b"1234565677")

with open("data.bin" ,'rb') as f:
   print("Binary read: " ,f.read())
   f.seek(6)
   print("After seek:",f.read())

```

# **6. SERIALIZATION AND DESERIALIZATION USING pickle**
Definition: Serialization = Convert object to bytes to save in file.

Deserialization = Convert bytes back to object.

```python
import pickle

data = {"name":"Ali" ,"marls ":[90,80,70]}

# serilization :object ->bytes->file
with open("data.pkl" ,'wb') as f:
  pickle.dump(data,f) #dump conver obj to bytes and write to file
  print("Data serilization and saved ")

  ## DESERIALIZATION: file -> bytes -> object
with open("data.pkl", "rb") as f: # Read binary
    loaded_data = pickle.load(f) # load() converts bytes back to object
    print("Data deserialized:", loaded_data) # Print recovered object
```

# **7. CUSTOM OBJECT PICKLING**
You can pickle your own class objects too.


```python
import pickle
class Student:
  def __init__(self, name, roll_no):
    self.name=name
    self.roll_no = roll_no
  def display(self):
    print(f"Name :{self.name} ,\nROllNo:{self.roll_no}")

s1 =Student("sara",101)

# pickele the object
with open ("student.pkl" , 'wb')as f :
  pickle.dump(s1,f)  #save obj to file

# Unpickle the object
with open("student.pkl",'rb') as f:
  s2= pickle.load(f)
  s2.display()

```

# **8. DECORATOR**
decoorator is a function that takes another function and add extra functionality

```python
def my_decorator(func):
  def wrapper():
    print("Befor function call:")
    func()
    print("After function call")

  return wrapper

@my_decorator
def greet():
  print("Hello!")


greet()
```

# **9. NAMESPACES, SCOPES, AND LEGB RULE**
Definition: LEGB = Local, Enclosing, Global, Built-in. Python searches variable in this order

```python
x = "golbal"
def outer():

  x = "Encloseing" #varriable in outer function
  def inner():
    x = "Local" #varrilbe in local scope
    print("Local x:",x)

    print("BuiltIn len:" , len("test"))

  inner()
  print("Enclosing x :", x)
outer()
print("Global x: ",x)

```
