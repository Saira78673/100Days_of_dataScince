# **Mini_project_2:Student result managment system**

```python
# ============================================
# Student Result Management System
# ============================================

# Function 1
# This function asks the user for student information.
def get_student():
    # Ask for student's name
    name = input("Enter Student Name: ")

    # Ask for marks in three subjects
    english = int(input("English Marks: "))
    math = int(input("Math Marks: "))
    science = int(input("Science Marks: "))

    # Return all values
    return name, english, math, science


# Function 2
# This function calculates total marks.
def calculate_total(english, math, science):

    # Add all marks
    total = english + math + science

    # Return total
    return total


# Function 3
# This function calculates percentage.
def calculate_percentage(total):

    # Divide by total marks
    percentage = (total / 300) * 100

    return percentage


# Function 4
# Decide grade according to percentage.
def find_grade(percentage):

    if percentage >= 90:
        return "A+"

    elif percentage >= 80:
        return "A"

    elif percentage >= 70:
        return "B"

    elif percentage >= 60:
        return "C"

    elif percentage >= 50:
        return "D"

    else:
        return "Fail"


# Function 5
# Display all results.
def display_result(name, total, percentage, grade):

    print("\n===== RESULT =====")

    print("Name :", name)

    print("Total Marks :", total)

    print("Percentage :", percentage)

    print("Grade :", grade)


# ==========================
# Main Program Starts Here
# ==========================

# Get all student data
name, english, math, science = get_student()

# Calculate total
total = calculate_total(english, math, science)

# Calculate percentage
percentage = calculate_percentage(total)

# Find grade
grade = find_grade(percentage)

# Display everything
display_result(name, total, percentage, grade)
```
